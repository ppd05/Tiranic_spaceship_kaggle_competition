```python
# Age distribution by Transported
sns.histplot(data=train, x='Age', hue='Transported', bins=30, kde=True, multiple='stack')
plt.title('Age Distribution by Transported')
plt.show()

# Distribution of spending categories by Transported
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in spending_cols:
    sns.histplot(data=train, x=col, hue='Transported', bins=30, kde=True, multiple='stack')
    plt.title(f'{col} Distribution by Transported')
    plt.show()
```

```python
# 1. Drop Unnecessary Columns
train = train.drop(['PassengerId', 'Name'], axis=1)

# 2. Extract Features from 'Cabin'
train[['Deck', 'CabinNum', 'Side']] = train['Cabin'].str.split('/', expand=True)
train = train.drop('Cabin', axis=1)

# 3. Handling Categorical Columns (fill with 'Unknown' and encode as category)
for col in ['HomePlanet', 'Destination', 'Deck', 'Side']:
    train[col] = train[col].fillna('Unknown').astype('category')

# 4. Robustly Convert Boolean Columns to 0/1 Integers
for col in ['CryoSleep', 'VIP']:
    train[col] = train[col].map({True: 1, False: 0, 'True': 1, 'False': 0})
    train[col] = train[col].fillna(train[col].mode()[0]).astype(int)

# 5. Fill Missing Numerical Columns with 0
num_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in num_cols:
    train[col] = train[col].fillna(0)

# 6. Set Spending Columns to 0 for Passengers in CryoSleep
spend_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
train.loc[train['CryoSleep'] == 1, spend_cols] = 0

# 7. Aggregate a Total Spending Feature
train['TotalSpend'] = train[spend_cols].sum(axis=1)

# 8. Discretize Age into Groups
train['AgeGroup'] = pd.cut(
    train['Age'],
    bins=[-1, 12, 18, 35, 55, 100],
    labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior']
)

# 9. Encode Categorical Features (excluding target)
train = pd.get_dummies(train, columns=['HomePlanet', 'Destination', 'Deck', 'Side', 'AgeGroup'],dtype=int)

# 10. Convert Target to 0/1 as Integer
train['Transported'] = train['Transported'].map({True: 1, False: 0, 'True': 1, 'False': 0}).astype(int)
```
